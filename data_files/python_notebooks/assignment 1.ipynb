```python
distance_in_five = [10.4, 11.0, 10.9, 11.0, 10.5]
distance_in_ten = [19.4, 17.6, 17.7, 16.5, 19.1]
distance_in_fifteen = [23.9, 26.6, 27.0, 23.5, 25.5]
degree_five = []
degree_ten = [5, 4, -5, -2, 0]
degree_fifteen = [-9, -6, -10, 10, -14]
```

```python
left_degree = [57, 56, 113, 59, 85]
left_time = [22.68, 22.53, 21.70, 22.30, 22.36]
right_degree = [80, 85, 97, 96, 104]
right_time = [18.50, 18.49, 18.06, 18.28, 18.50]
around_degree = [172, 177, 127, 176, 134]
around_time = [32.35, 32.17, 31.70, 32.23, 32.15]
```

```python
feedback_five = [7.1, 4.0, 4.5, 4.2, 5.7]
feedback_five_wall = [18.9, 18.5, 18.6, 18.4, 19.2]
fb_ten = [10.5, 10.6, 11.3, 10.9, 10.4]
fb_ten_wall = [20.1, 20.5, 19.5, 20.2, 20.9]
fb_fifteen = [17.9, 18.6, 17.1, 16.8, 18.6]
fb_fifteen_wall = [20.6, 22.5, 19.2, 21.4, 21.5]
```

```python
cosine_ten = []

for i in degree_ten:
    add = math.cos(math.radians(i))
    cosine_ten.append(add)

combined = zip(cosine_ten, distance_in_ten)

forward_distance_ten = []

for a, b in combined:
    new = abs(a) * b
    forward_distance_ten.append(new)
    
forward_distance_ten
```

```python
cosine_fifteen = []

for i in degree_fifteen:
    add = math.cos(math.radians(i))
    cosine_fifteen.append(add)

combined = zip(cosine_fifteen, distance_in_fifteen)

forward_distance_fifteen = []

for a, b in combined:
    new = abs(a) * b
    forward_distance_fifteen.append(new)
    
forward_distance_fifteen
```

```python
diff_ten = []

for a,b in zip(distance_in_ten, forward_distance_ten):
    add = a - b
    diff_ten.append(add)
    
diff_ten
```

```python
diff_fifteen = []

for a,b in zip(distance_in_fifteen, forward_distance_fifteen):
    add = a - b
    diff_fifteen.append(add)
    
diff_fifteen
```

```python
#WALKING GAIT DIFFERENCES

#Building our plots
with plt.style.context('fivethirtyeight'):
    #initialize figure
    fig = plt.figure(figsize= (14,8))

    #first axis
    ax1 = fig.add_subplot(2, 2, 1)
    ax1.scatter(trials, diff_ten, color="steelblue", s=80)
    ax1.set_xlabel("Trials", fontsize=15)
    ax1.set_ylabel("Difference for 10$s$ $(cm)$", fontsize=15)
    
    #adjust y-axis
    ax1.axis(xmin=0, xmax=5.9, ymin=-0.01, ymax=0.3)
    
    #more adjusting
    for axis in ['bottom','left']:
        ax1.spines[axis].set_linewidth(0.5)
        ax1.spines[axis].set_position(("axes", -0.05))
    
    #add a title
    plt.suptitle('Difference between Total Distance Walked and Straight Forward Distance $(cm)$', fontsize=20)
    
    #second axis
    ax2 = fig.add_subplot(2, 2, 2)
    ax2.scatter(trials, diff_fifteen, color="steelblue", s=80)
    ax2.set_xlabel("Trials", fontsize=15)
    ax2.set_ylabel("Difference for 15$s$ $(cm)$", fontsize=15)
    
    #adjust second plot axes
    ax2.axis(xmin=0, xmax=5.9, ymin=-.1, ymax=1)
    
    #more adjusting
    for axis in ['bottom','left']:
        ax2.spines[axis].set_linewidth(0.5)
        ax2.spines[axis].set_position(("axes", -0.05))
    
    #saving the graph
    plt.savefig('difference_10_and_15s.png')
```
